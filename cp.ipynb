```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dp = pd.read_csv("cure_a_princesa.csv")
dp.info()
```

```python
dp_sem_nan = dp.dropna()
dp_sem_nan.info()
```

```python

df_curada = dp_sem_nan['Curada'] == 1
df_nao_curada = dp_sem_nan['Curada'] == 0

df_curada = dp_sem_nan[df_curada]
df_nao_curada = dp_sem_nan[df_nao_curada]

display(df_curada) # coisas que curaram a princesa
display(df_nao_curada) # coisas que não curaram a princesa

```

```python
# Filtrando os dados
df_curada_presa_0 = df_curada[df_curada['Presa de quimera'] == 0.0]
df_nao_curada_presa_not_0 = df_nao_curada[df_nao_curada['Presa de quimera'] != 0.0]

# Criando o gráfico para df_curada_presa_0
sns.kdeplot(x='Presa de quimera', data=df_curada_presa_0)
plt.title('Contagem de "Presa de quimera" igual a 0.0 em df_curada')
plt.show()

# Criando o gráfico para df_nao_curada_presa_not_0
sns.kdeplot(x='Presa de quimera', data=df_nao_curada_presa_not_0)
plt.title('Contagem de "Presa de quimera" diferente de 0.0 em df_nao_curada')
plt.show()
```

```python
# um gráfico descritivo que mostra como está a distribuição de ingredientes que curaram a princesa na tabela de curada
df_curada.describe()
```

```python
df_curada_copy = df_curada.copy()
df_nao_curada_copy = df_nao_curada.copy()

df_curada_copy.drop(columns='Curada').describe().plot(kind='bar')
plt.title('Distribuição de ingredientes que curaram a princesa')
plt.show()

df_nao_curada_copy.drop(columns='Curada').describe().plot(kind='bar')
plt.title('Distribuição de ingredientes que não curaram a princesa')
plt.show()

```

```python
# Calculando a correlação
correlation = df_curada.corr()

# Criando o gráfico de calor
sns.heatmap(correlation, annot=True, cmap='coolwarm')

# Adicionando título
plt.title('Correlação entre Ingredientes')

# Mostrando o gráfico
plt.show()
```

```python
df_curada_filtrada = df_curada[df_curada > 0]

# Calculando a correlação
correlation = df_curada_filtrada.corr()

# Criando o gráfico de calor
sns.heatmap(correlation, annot=True, cmap='coolwarm')

# Adicionando título
plt.title('Correlação entre Ingredientes com Quantidade Acima de 0')

# Mostrando o gráfico
plt.show()
```
